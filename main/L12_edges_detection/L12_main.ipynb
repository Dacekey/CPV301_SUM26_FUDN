```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = "tiger_image.jpg"
	
# Read the original image
img = cv2.imread(img_path,flags=0) 
# Blur the image for better edge detection
# img_blur = cv2.GaussianBlur(img,(3,3), sigmaX=0, sigmaY=0) 
img_blur = cv2.GaussianBlur(img,(3,3), 0) 
```

```python
# Show ảnh gốc và ảnh blur
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap="gray")
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img_blur, cmap="gray")
plt.title("Gaussian Blur")
plt.axis("off")

plt.show()
```

```python
# Sobel Edge Detection
sobelx = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
```

```python
# Using abs to normalize for easily looking because of dark.
sobelx_show = cv2.convertScaleAbs(sobelx)
sobely_show = cv2.convertScaleAbs(sobely)
sobelxy_show = cv2.convertScaleAbs(sobelxy)
```

```python
# Display Sobel Edge Detection Images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(sobelx_show, cmap="gray")
plt.title("Sobel X")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(sobely_show, cmap="gray")
plt.title("Sobel Y")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(sobelxy_show, cmap="gray")
plt.title("Sobel X Y")
plt.axis("off")

plt.show()
```

```python
# Canny Edge Detection
canny_test_1 = cv2.Canny(image=img_blur, threshold1=100, threshold2=200) 
canny_test_2 = cv2.Canny(image=img_blur, threshold1=30, threshold2=100) 
```

```python
# Display Canny Edge Detection Image
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(canny_test_1, cmap="gray")
plt.title("Canny: min_thrs=100, max_thrs=200")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(canny_test_2, cmap="gray")
plt.title("Canny: min_thrs=30, max_thrs=100")
plt.axis("off")

plt.show()
```
